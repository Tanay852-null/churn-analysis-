```python
## Univariate Analysis (One Variable at a Time)
```

```python
sns.countplot(data=df, x="Churn", hue="Churn", palette="coolwarm", legend=False)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.histplot(df["MonthlyCharges"], bins=30, kde=True)
plt.title("Distribution of Monthly Charges")
plt.show()
```

Helps identify whether the charges are normally distributed or crooked



```python
##Bivariate Analysis (Two Variables Relationship)
```

```python
sns.boxplot(data=df, x="Churn", y="tenure", hue="Churn", palette="coolwarm", dodge=False)
plt.title("Tenure vs Churn")
plt.show()
```

If customers with lower tenure churn more, retention strategies should focus on new customers.

```python
sns.kdeplot(df[df["Churn"] == "Yes"]["MonthlyCharges"], label="Churn", fill=True, color="red")
sns.kdeplot(df[df["Churn"] == "No"]["MonthlyCharges"], label="No Churn", fill=True, color="blue")
plt.title("Monthly Charges vs Churn")
plt.legend()
plt.show()
```

If higher Monthly Charges are linked to churn, pricing strategies may need adjustments.

```python
corr_matrix = df.select_dtypes(include=["number"]).corr()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

Identifies highly correlated features, helping in feature selection.

```python
##Multivariate Analysis (More Than Two Variables)
```

```python
sns.pairplot(df[["tenure", "MonthlyCharges", "TotalCharges", "Churn"]], hue="Churn", palette="coolwarm")
plt.show()
```

Useful for understanding the combined effect of multiple features on churn.

```python
## Univariate Analysis → Understands individual feature distributions (Churn rate, Monthly Charges, etc.).

## Bivariate Analysis → Finds relationships between two variables (Tenure vs Churn, Monthly Charges vs Churn).

## Multivariate Analysis → Examines interactions between multiple features (Pairplots, Correlation Matrix).
```
